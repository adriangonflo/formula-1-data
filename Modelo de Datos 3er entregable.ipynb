```python
import requests
import json
import pandas as pd 
import warnings
import numpy as np
warnings.filterwarnings("ignore")
```

# Tabla drverStandigs

```python
url = 'http://ergast.com/api/f1/2008/driverStandings.json'
response = requests.get(url)
content=json.loads(response.content)
```

```python
content['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings']
```

```python
drst={'DriverID':[],'Points':[],'Positions':[],'PositionText':[],'wins':[]}
for year in list(range(2008,2023)):
    url='http://ergast.com/api/f1/{}/driverStandings.json'
    response = requests.get(url.format(year))
    content=json.loads(response.content)
    for item in content['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings']:
        drst['DriverID'].append(item['Driver']['driverId'])
        drst['Points'].append(item['points'])
        drst['Positions'].append(item['position'])
        drst['PositionText'].append(item['positionText'])
        drst['wins'].append(item['wins'])
pd.DataFrame(drst).to_csv('driversstanding.csv',index=False)
```

```python
url = 'http://ergast.com/api/f1/2008/constructorStandings.json'
response = requests.get(url)
content=json.loads(response.content)
```

```python
content['MRData']['StandingsTable']['StandingsLists'][0]['ConstructorStandings']
```

```python
cost={'constructorId':[],'Position':[],'PositionTex':[],'Points':[],'wins':[]}
for year in list(range(2008,2023)):
    url='http://ergast.com/api/f1/{}/constructorStandings.json'
    response = requests.get(url.format(year))
    content=json.loads(response.content)
    for item in content['MRData']['StandingsTable']['StandingsLists'][0]['ConstructorStandings'] :
        cost['constructorId'].append(item['Constructor']['constructorId'])
        cost['Points'].append(item['points'])
        cost['Position'].append(item['position'])
        cost['PositionTex'].append(item['positionText'])
        cost['wins'].append(item['wins'])
pd.DataFrame(cost).to_csv('constructorstanding.csv',index=False)
```

```python
url='http://ergast.com/api/f1/2011/5/laps/1.json'
response = requests.get(url)
content=json.loads(response.content)
```

```python
content['MRData']['RaceTable']['Races'][0]['Laps']
```

```python
lapt={'driverid':[],'lap':[],'Position':[],'Time':[]}
for year in list(range(1996,2023)):
    url='http://ergast.com/api/f1/{}/5/laps/1.json'
    response = requests.get(url.format(year))
    content=json.loads(response.content)
    for item in content['MRData']['RaceTable']['Races'][0]['Laps'][0]['Timings']:
        lapt['driverid'].append(item['driverId'])
        lapt['Position'].append(item['position'])
        lapt['Time'].append(item['time'])
        lapt['lap'].append(content['MRData']['RaceTable']['Races'][0]['Laps'][0]['number'])
pd.DataFrame(lapt).to_csv('LapTimes.csv', index=False)
```

# Tabal de constructorResults

```python
conres = {'constructorId': [], 'points': [], 'status': []}
for year in list(range(2010, 2023)):
    url = "http://ergast.com/api/f1/{}/results.json?limit=1000"
    response = requests.get(url.format(year))
    content = json.loads(response.content)
    for item in content['MRData']['RaceTable']['Races']:
        for item2 in item['Results']:
            conres['constructorId'].append(item2['Constructor']['constructorId'])
            conres['points'].append(float(item2['points']))
            conres['status'].append(item2['status'])
pd.DataFrame(conres).to_csv('constructorResults.csv', index=False )
```
